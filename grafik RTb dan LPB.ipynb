```python
from math import log10, sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
#-- kabel FO G.652 - SI-SM
L=np.arange(1,5.422,1) #km
Hspl=2 #km
disp_M=3.56e-12 #s/km.nm
alpha_f=0.3 #dB/km
```

```python
#-- Pengirim
Ptx_db=5 #dBm
spektral=1 #nm
t_tx=0.16e-12 #s
Mod_NRZ=0.7 #NRZ
Mod_RZ=0.35 #RZ
Br=2.4e9 #2.4Gbps
Tb=1/Br #s
```

```python
#--Pengirim
Sensitivitas = -28 #dBm
t_rx=0.16e-12 #s
```

```python
#-- Redaman Kanal
N_c=2
alpha_c=0.25 #dB/titik
N_sp=(L/Hspl)-1
alpha_sp=0.05 #dB/titik
Ms=4 #dB
```

## Link Power Budget
$$\alpha_T= 2\alpha_c+N_{sp}\alpha_{sp}+L\alpha_f+M_s$$

```python
alpha_t=N_c*alpha_c+N_sp*alpha_sp+L*alpha_f+Ms
print ("Redaman Kabel ",alpha_t, "dB")
```

```python
Prx=Ptx_db-alpha_t
```

```python
Prx
```

```python
## Rise Time Budget
$$t_(sys)+\sqrt{t_{tx}^2+t_{mod}^2+t_{wg}^2+t_{rx}}. $$
```

```python
t_mat=disp_M*L*spektral
t_wg=0 #diabaikan
t_mod=0 #SM

t_sys=np.sqrt(t_tx**2+
          t_mat**2+
          t_mod**2+
          t_wg**2+
           t_rx)
```

```python
t_sys
```

```python
plt.plot(L,Prx)
plt.xlabel("Jarak(km)")
plt.ylabel("Daya Terima (dBm)")
plt.title("Kurva Link Power Budget")
```

```python
plt.plot(L,t_sys)
plt.xlabel("Jarak(km)")
plt.ylabel("Dispersi sistem(s)")
plt.title("Kurva Rise Time Budget")
```
